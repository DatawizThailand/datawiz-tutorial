# Interpolations for imshow

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']

np.random.seed(19680801)

grid = np.random.rand(4, 4)

fig, axes = plt.subplots(6, 6, figsize=(14, 14), subplot_kw={'xticks':[], 'yticks':[]})
fig.subplots_adjust(hspace=0.3, wspace=0.05)

colors = ['viridis', 'seismic']

for ax, obj in zip(axes.flat, itertools.product(colors, methods)):
    color, interp_method = obj
    ax.imshow(grid, interpolation=interp_method, cmap=color)
    ax.set_title(interp_method, fontsize=18)

plt.show()
```

### MNIST

```python
img = plt.imread('docs/mnist.png')

fig, axes = plt.subplots(3, 6, figsize=(14, 8), subplot_kw={'xticks':[], 'yticks':[]})
fig.subplots_adjust(hspace=0.3, wspace=0.05)

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(img, interpolation=interp_method)
    ax.set_title(interp_method, fontsize=18)
plt.show()
```
