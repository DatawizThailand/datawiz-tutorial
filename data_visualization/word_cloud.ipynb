```python
from wordcloud import WordCloud

import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
text = open(os.path.join('../natural_language_processing/data/constitution.txt')).read()

wordcloud = WordCloud(
    background_color='white',
    colormap='Blues', 
    width=1024,
    height=768
).generate(text)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
